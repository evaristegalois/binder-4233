<h4>The Skydiver</h4>

A skydiver with a mass of $60\,kg$ jumps from a helicopter
  hovering at an initial altitude of $5$ kilometers. Assume that she
  falls vertically with initial velocity zero and experiences an
  upward force $F_{R}$ of air resistance given in terms of her velocity $v$
  (in metres per second) by
  $$  F_{R}=(0.0096)(100v+10v^{2}+v^{3})\notag
$$
(in Newtons and with the coordinate axis directed downward so that
$v>0$ during her descent to the ground). If she does not open her
parachute, what will be her terminal velocity? How fast will she be
falling after five seconds have elapsed? After ten seconds? After 20
seconds?

```julia
k=0.0096
```

```julia
g=9.81
```

```julia
f(x,y)=g-(k/60)*(100*y+10*y^2+y^3)
```

```julia
using LinearAlgebra
```

```julia
using Plots
```

```julia
h=0.1
```

```julia
Ec=[0 0]
```

```julia
for i in 1:600
    Ec=vcat(Ec,[Ec[i,1]+h Ec[i,2]+(h/6)*(f(Ec[i,1],Ec[i,2])+2*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2]))+2*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2])))+f(Ec[i,1]+h,Ec[i,2]+h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2])))))])
end
```

```julia
plot(Ec[:,1],Ec[:,2],ylims=(-1,50),title="The Skydiver",lw=3)
```
