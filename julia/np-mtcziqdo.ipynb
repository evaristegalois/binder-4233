\begin{align*}

```julia
M2=[sol[1,1] 0.3 # filling in initial conditions
sol[2,1] 0.3]
```

```julia
G1=(1/20)*inv(M2)
```

```julia
M3=[1 1
G1[1,1]+G1[2,1] G1[1,2]+G1[2,2]]
```

```julia
v3=[0.5
0.5]
```

```julia
inv(M3)*v3
```

```julia
using DifferentialEquations
```

```julia
function brine(du,u,p,t)
    du[1]=p[1]*u[1]+p[2]*u[2]
    du[2]=p[3]*u[1]+p[4]*u[2]
end
```

```julia
u0=[0.5;0.5]
```

```julia
p=(-0.3,0.05,0.3,-0.15)
```

```julia
tspan=(0.0,40.0)
```

```julia
prob=ODEProblem(brine,u0,tspan,p)
```

```julia
solution=solve(prob)
```

```julia
using Plots
```

```julia
plot(solution,lw=2)
```
