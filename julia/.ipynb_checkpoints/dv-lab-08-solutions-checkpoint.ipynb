<h3>
    Lab Sheet 8 Problem 6
</h3>

Find a general solution for
\begin{align*}
  x'&=y+z+e^{-t}\\
  y'&=x+z\\
  z'&=x+y
\end{align*}

$$
\left[
\begin{array}{ccc}
D&-1&-1\\-1&D&-1\\-1&-1&D
\end{array}
\right]\vec{x}=\left[
\begin{array}{c}
e^{-t}\\0\\0\end{array}
\right]
$$

```julia
using SymPy
```

```julia
@vars D t
```

```julia
using LinearAlgebra
```

```julia
A6=[D -1 -1
-1 D -1
-1 -1 D]
```

```julia
v6=[exp(-t);0;0]
```

```julia
inv(A6)*v6
```

Consequently,
$$
\begin{align*}
x(t)&=\frac{2}{3}te^{-t}+a_1e^{-t}+a_2e^{2t}\\
y(t)&=-\frac{1}{3}te^{-t}+b_1e^{-t}+b_2e^{2t}\\
z(t)&=-\frac{1}{3}te^{-t}+c_1e^{-t}+c_2e^{2t}\\
\end{align*}
$$
Plug into system to find
$$
\begin{align*}
b_2&=a_2\\
c_2&=a_2\\
c_1&=-a_1-b_1-\frac{1}{3}
\end{align*}
$$
for three constants as you would expect for a first-order system with three variable functions.

<h3>
    Lab Sheet 8 Problem 7
</h3>

Suppose that an artillery projectile is fired from ground
level with initial velocity $3000\,ft/s$ and initial inclination angle
$40^{\circ}$. Assume that its air resistance deceleration is
$(0.0001)v^2$.
* What is the range of the projectile and what is its total time of
flight? What is its speed at impact with the ground?
* What is the
maximum altitude of the projectile, and when is that altitude
attained?
* You will find that the projectile is still losing speed
at the apex of its trajectory. What is the minimum speed that it
attains during its descent?

Let $a=0.0001,x_1(t)=x(t),x_2(t)=x'(t),x_3(t)=y(t),x_4(t)=y'(t),g=32$. Let $\theta$ be such that $\cos\theta=\frac{x_2}{\sqrt{x_2^2+x_4^2}}$ and $\sin\theta=\frac{x_4}{\sqrt{x_2^2+x_4^2}}$. Then the system is
$$
\begin{align*}
x_1'&=x_2\\
x_2'&=-\cos\theta\cdot{}a(x_2^2+x_4^2)\\
x_3'&=x_4\\
x_4'&=-\sin\theta\cdot{}a(x_2^2+x_4^2)\cdot\frac{x4}{|x4|}-g
\end{align*}
$$
with 
$$
\begin{align*}
x_1(0)&=0\\
x_2(0)&=3000\cos(40^{\circ})\\
x_3(0)&=0\\
x_4(0)&=3000\sin(40^{\circ})\\
\end{align*}
$$


```julia
using DifferentialEquations
using Plots
using ParameterizedFunctions
```

```julia
g=32
```

```julia
vox=*(3000,cos(*(40,/(pi,180))))
```

```julia
voy=*(3000,sin(*(40,/(pi,180))))
```

Let $a=0.0001,x_1(t)=x(t),x_2(t)=x'(t),x_3(t)=y(t),x_4(t)=y'(t),g=32$. Let $\theta$ be such that $\cos\theta=\frac{x_2}{\sqrt{x_2^2+x_4^2}}$ and $\sin\theta=\frac{x_4}{\sqrt{x_2^2+x_4^2}}$. Then the system is
$$
\begin{align*}
x_1'&=x_2\\
x_2'&=-\cos\theta\cdot{}a(x_2^2+x_4^2)\\
x_3'&=x_4\\
x_4'&=-\sin\theta\cdot{}a(x_2^2+x_4^2)\cdot\frac{x4}{|x4|}-g
\end{align*}
$$
with 
$$
\begin{align*}
x_1(0)&=0\\
x_2(0)&=3000\cos(40^{\circ})\\
x_3(0)&=0\\
x_4(0)&=3000\sin(40^{\circ})\\
\end{align*}
$$


```julia
pt! = @ode_def ProjectileTrajectory begin
    dx1=x2
    dx2=*(-a,+(^(x2,2),^(x4,2)),/(x2,sqrt(+(^(x2,2),^(x4,2)))))
    dx3=x4
    dx4=-(*(-a,+(^(x2,2),^(x4,2)),/(x4,sqrt(+(^(x2,2),^(x4,2))))),g)
    end a
u0=[0,vox,0,voy]
p=(0.0001)
tspan=(0,48)
problem=ODEProblem(pt!,u0,tspan,p)
solution=solve(problem,saveat=0.01)
plot(solution,lw=2)
```

```julia
plot(solution,vars=(1,3),xaxis="horizontal displacement",yaxis="vertical displacement")
```

Results (all approximate): range $21416\,ft$, total time 45.96
seconds, speed at impact $518\,ft/s$; maximum altitude $8969\,ft$,
apex time $17.56\,s$; minimum speed $368\,ft/s$ at $t\approx{}23$.
