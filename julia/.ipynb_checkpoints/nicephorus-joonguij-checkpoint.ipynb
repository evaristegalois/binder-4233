<h4>Euler's Method</h4>

Apply Euler's method to approximate the solution of the initial value problem
$$
\frac{dy}{dx}=x+\frac{1}{5}y,\qquad{}y(0)=-3
$$

The solution is easily found using LFE.
$$
y(x)=22e^{\frac{1}{5}x}-5x-25
$$

```julia
f(x,y)=x+(1/5)*y
```

```julia
using LinearAlgebra
```

```julia
Ea=[0 -3]
```

```julia
for i in 1:5
    Ea=vcat(Ea,[Ea[i,1]+1 Ea[i,2]+f(Ea[i,1],Ea[i,2])])
end
```

```julia
Ea
```

```julia
using Plots
```

```julia
x=range(0,stop=5,length=100)
```

<h4>
    Improved Euler Method
</h4>

```julia
Eb=[0 -3]
```

```julia
for i in 1:5
    Eb=vcat(Eb,[Eb[i,1]+1 Eb[i,2]+((1/2)*(f(Eb[i,1],Eb[i,2])+f(Eb[i,1]+1,Eb[i,2]+f(Eb[i,1],Eb[i,2]))))])
end
```

```julia
Eb
```

```julia
plot(Ea[:,1],Ea[:,2],ylims=(-5,10),title="Euler's Method",label=["Euler's Method"],lw=3) # my Euler approximation
plot!(x,(22 .* exp.((1/5) .* x)) .- ( 5 .* x ) .- 25,label=["Algebraic Solution"],lw=3) # the algebraic solution
plot!(Eb[:,1],Eb[:,2],label=["Improved Euler Method"],lw=3) # my improved Euler approximation
```

<h4>
    Runge-Kutta Method
</h4>

```julia
f(x,y)=x+y
```

```julia
h=0.1
```

```julia
Ec=[0 1]
```

```julia
x=range(0,stop=1,length=100)
```

```julia
for i in 1:10
    Ec=vcat(Ec,[Ec[i,1]+h Ec[i,2]+(h/6)*(f(Ec[i,1],Ec[i,2])+2*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2]))+2*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2])))+f(Ec[i,1]+h,Ec[i,2]+h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2])))))])
end
```

```julia
Ec
```

```julia
plot(Ec[:,1],Ec[:,2],ylims=(-1,4),title="Runge-Kutta Method",label=["Runge-Kutta Method"],lw=3) # my Runge-Kutta approximation
plot!(x,2 .* exp.(x) .- x .- 0.9,label=["Algebraic Solution plus 1/10"],lw=3) # the algebraic solution
```
