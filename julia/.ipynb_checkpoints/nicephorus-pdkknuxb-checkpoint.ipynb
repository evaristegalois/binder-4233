<h4>Assignment 2 Problem 2: Bobbing Barrel</h4>

A barrel is dropped into the ocean, striking the surface of the water in an upright position
at time $t=0$ with a downward velocity $v=0.67\,m/s$. In addition to the downward force
of gravity, the barrel is acted upon by a force of viscous resistance proportional to its
velocity, and an upward buoyancy force equal to the weight of the water displaced by
the barrel.

Model the barrel as a cylinder of radius $R=0.40\,m$
and height $H=0.98\,m$, and assume that its density
$\rho=622\,kg/m^{3}$ is approximately constant. Take the
density of seawater to be $\rho_{w}=1024\,kg/m^{3}$, and the
coefficient of linear resistance appropriate for
movement through water to be $c=297\,kg/s$.
Assuming that the barrel remains always upright,
find the depth $x$ of the bottom of the barrel as a
function of time $t$. Make sure to indicate what
coordinate system you are using (positive up or
positive down).

Use Maple or Excel to plot the depth $x$ for the first $5$ seconds OR until the barrel
becomes completely submerged, if that occurs before $5$ seconds have elapsed. If you
use Excel, use a time increment of $0.1$ seconds (unless the barrel becomes completely
submerged before $5$ seconds have elapsed, in which case you should use a time increment
of $0.01$ seconds).

Why would your solution become invalid if the barrel were to become completely
submerged? Make reference to the differential equation in your answer.

First of all, some constants. Using a coordinate system where $x$ increases as the barrel goes down.

```julia
R=0.4
```

```julia
H=0.98
```

```julia
rho=622
```

```julia
m=R^2*pi*H*rho
```

```julia
rhow=1024
```

```julia
k=rhow*R^2*pi
```

```julia
c=297
```

```julia
v0=0.67
```

```julia
g=9.8
```

```julia
using Polynomials
```

This is the differential equation:
$$
mx''+cx'+kx=g
$$

```julia
p5=Polynomial([k,c,m])
```

```julia
roots(p5)
```

The solution for the associated homogeneous differential equation is
$$
x_{c}(t)=e^{-0.48t}\left(c_{1}\cos(1.20t)+c_{2}\sin(1.20t)\right)
$$
$x_{p}=\frac{g}{k}$ is obvious, so
$$
x(t)=0.019+e^{-0.48t}\left(c_{1}\cos(1.20t)+c_{2}\sin(1.20t)\right)
$$

```julia
phi=imag(roots(p5)[2,1])
```

```julia
psi=real(roots(p5)[2,1])
```

```julia
A=g/k
```

Now plug in $x(0)=0$ and $x'(0)=v_{0}$ for

```julia
c1=-A
```

```julia
c2=(v0-psi*c1)/phi
```

```julia
using Plots
```

```julia
rng=range(-0.2,stop=12,length=500)
```

```julia
x(t)=A+exp(psi*t)*(c1*cos(phi*t)+c2*sin(phi*t))
```

```julia
plot(rng,x.(rng),line=2,leg=false)
```

We are lucky. The barrel is $0.98\,m$ tall and will not be submerged. If it were, we would have to change the differential equation as the $514.7x$ term would no longer depend on $x$ and become constant.
