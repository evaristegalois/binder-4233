<h5>
    Lab Sheet 5 Problem 1
</h5>

A skydiver with a mass of $60\,kg$ jumps from a helicopter
  hovering at an initial altitude of $5$ kilometers. Assume that she
  falls vertically with initial velocity zero and experiences an
  upward force $F_{R}$ of air resistance given in terms of her velocity $v$
  (in metres per second) by
  $$  F_{R}=(0.0096)(100v+10v^{2}+v^{3})\notag
$$
(in Newtons and with the coordinate axis directed downward so that
$v>0$ during her descent to the ground). If she does not open her
parachute, what will be her terminal velocity? How fast will she be
falling after five seconds have elapsed? After ten seconds? After 20
seconds?

```julia
k=0.0096
```

```julia
g=9.81
```

The differential equation for the position function $u(t)$ is
$$
mu''+k(100u'+10(u')^{2}+(u')^{3})=-mg
$$
now substitute $y=u'$ for the skydiver's velocity.

```julia
f(t,y)=g-(k/60)*(100*y+10*y^2+y^3)
```

```julia
using LinearAlgebra
```

```julia
using Plots
```

```julia
h=0.1
```

```julia
Ec=[0 0] # implementing Runge-Kutta in Julia
```

```julia
for i in 1:600
    Ec=vcat(Ec,[Ec[i,1]+h Ec[i,2]+(h/6)*(f(Ec[i,1],Ec[i,2])+2*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2]))+2*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2])))+f(Ec[i,1]+h,Ec[i,2]+h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1]+0.5*h,Ec[i,2]+0.5*h*f(Ec[i,1],Ec[i,2])))))])
end
```

```julia
plot(Ec[:,1],Ec[:,2],ylims=(-1,50),title="The Skydiver",lw=3)
```

```julia
Ec
```

<h5>
    Lab Sheet 5 Problem 3
</h5>

Find the solution set for the following system of linear
  equations. (Hint: use a computer algebra system and use
  nonhomogeneous and associated homogeneous systems.)
$$
    \begin{array}{ccccccccccc}
  5x & + & 10y & + & 3z & - & w  & - & 5u & = & 4  \\
 -4x & - & 8y  & - & 3z & - & w  & + & 4u & = & -2 \\
  x  & + & y   & - & 4z & + & 3w &   &    & = & 9  \\
     &   & y   & + & 4z & - & 5w & - & u  & = & -7 \\
 -3x & - & 6y  & - & 2z &   &    & + & 3u & = & -2 \\
    \end{array}
$$

```julia
A1=[5 10 3 -1 -5
-4 -8 -3 -1 4
1 1 -4 3 0
0 1 4 -5 -1
-3 -6 -2 0 3
]
```

```julia
det(A1)
```

```julia
A2=[5 10 3 -1 -5 4
-4 -8 -3 -1 4 -2
1 1 -4 3 0 9
0 1 4 -5 -1 -7
-3 -6 -2 0 3 -2
]
```

```julia
using RowEchelon
```

```julia
A3=rref(A2)
```

The last two equations are redundant, and the rank is $3$. The nullity is $2$, therefore we need one solution $\vec{v}$ for the nonhomogeneous system of linear equations and two linearly independent solutions $\vec{b_{1}},\vec{b_{2}}$ for the homogeneous system. Let's begin by finding $\vec{v}$. The third equation $z+3w=-2$ is solved, for example, by $z=1,w=-1$ (it's a good idea to pick integer solutions at this point if you can). The second equation $y-u=-16$ is solved by, for example, $y=-8,u=8$. The first equation (always go from the bottom to the top) then determines $x=24$. Let's test this solution for the nonhomogeneous system.

```julia
v=[24;-8;1;-1;8]
```

```julia
A1*v # it works!
```

Now replace the last column of the row echelon form by all zeroes to turn the nonhomogeneous system into a homogeneous system. 


```julia
A4=hcat(A3[:,1:end.!=6],[0;0;0;0;0])
```

```julia
b1=[-24;9;-3;1;-8]
```

```julia
b2=[-1;1;0;0;1]
```

```julia
A1*b1 # it works!
```

```julia
A1*b2 # it works!
```

Therefore, the solution set is
$$
\left\{\vec{x}\in\mathbb{R}^{5}\left\vert\vec{x}=\left(\begin{array}{c}24\\-8\\1\\-1\\8\end{array}\right)+\lambda_{1}\cdot\left(\begin{array}{c}-24\\9\\-3\\1\\-8\end{array}\right)+\lambda_{2}\cdot\left(\begin{array}{c}-1\\1\\0\\0\\1\end{array}\right),\lambda_{i}\in\mathbb{R}\right.\right\}
$$

<h5>
    Lab Sheet 5 Problem 4
</h5>

A lunar lander is free-falling toward the moon, and at an altitude of $53$ kilometers above the
lunar surface its downward velocity is measured at $1477\,km/h$. Its retrorockets, when fired
in free space, provide a deceleration of $T=4\,m/s^{2}$. At what height above the lunar surface
should the retrorockets be activated to ensure a soft touchdown ($v=0$ at impact)?

$M$ is the mass of the moon. $R$ is the radius of the moon, $G$ is the gravitational constant.

```julia
M=7.35e22
```

```julia
R=1.74e6
```

```julia
G=6.6726e-11
```

```julia
v2=1477*1000/3600
```

```julia
t1=v2/4
```

```julia
(-2)*t1^2+410*t1
```

Without gravity, $s_{1}$ is about $21\,km$, the height at which the retrorockets should be fired. 

$$
\frac{d^{2}r}{dt^{2}}=T-\frac{GM}{r^{2}}
$$
implies
$$
\frac{1}{2}v^{2}=\frac{GM}{r}+Tr+C
$$
since $v(0)=0$,
$$
C_{1}=-\frac{GM}{R}-4R
$$

```julia
C1=((-G*M)/R)-4*R
```

and 
$$
\frac{1}{2}v^{2}=\frac{GM}{r}+4r+C_{1}
$$
after ignition. Before ignition, we know that
$$
\frac{1}{2}v_{2}^{2}=\frac{GM}{R+53000}+C_{2}$$

```julia
C2=0.5*v2^2-((G*M)/(R+53000))
```

At $t_{1}$, the two functions agree, so
$$
\frac{GM}{r(t_{1})}+C_{2}=\frac{GM}{r(t_{1})}+4r(t_{1})+C_{1}
$$
and therefore

```julia
r1=(C2-C1)/4
```

```julia
r1-R
```

The retrorockets should be fired at $41870$ metres above the lunar surface.
