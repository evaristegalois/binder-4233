Calculations in Schmierbuch 3620 (leonardo) and 3551 (hand).
$$
y_c(t)=c_1\left[\begin{array}{c}\cos{}t\\-\sin{}t\end{array}\right]+c_2\left[\begin{array}{c}\sin{}t\\\cos{}t\end{array}\right]
$$
$$
y_p(t)=\Phi(t)\int\left(\Phi(t)\right)^{-1}\left[\begin{array}{c}0\\\tan{}t\end{array}\right]\,dt
$$
The particular solution is
\begin{align}
y_p&=\cos{}t\cdot\ln|\sec{}t+\tan{}t| \\
y_p'&=-\sin{}t\cdot\ln|\sec{}t+\tan{}t|+1
\end{align}
This matches my result in `dv-whiteboard-20220223A.png` and `dv-whiteboard-20220330A.png`. For the initial condition $y(0)=0$ and $y'(0)=0$ we get $c_1=0,c_2=-1$.

```julia
using DifferentialEquations
using Plots
```

```julia
function sistem(du,u,p,t)
    du[1]=p[1]*u[1]+p[2]*u[2]+0
    du[2]=p[3]*u[1]+p[4]*u[2]+tan(t)
end
u0=[0;0]
p=(0,1,-1,0)
tspan=(0.0,6.0)
problem=ODEProblem(sistem,u0,tspan,p)
solution=solve(problem)
rng=range(0,stop=6,length=5000)
plot(solution,lw=2)
plot!(rng,-cos.(rng) .* log.(abs.(sec.(rng) .+ tan.(rng))),lw=2,lc=:green)
```

<h2>Runge Kutta Method</h2>

```julia
t=[0]
```

```julia
E=[0 0]
```

```julia
h=0.00001
```

```julia
f(t,x)=[x[1,2] -x[1,1]+tan(t[1,1])]
```

```julia
for i in 0:599999
    k1=f(t[i+1,1]+i*h,transpose(E[i+1,:]))
    k2=f(t[i+1,1]+(i+(1/2))*h,transpose(E[i+1,:])+(1/2)*k1*h)
    k3=f(t[i+1,1]+(i+(1/2))*h,transpose(E[i+1,:])+(1/2)*k2*h)
    k4=f(t[i+1,1]+(i+1)*h,transpose(E[i+1,:])+k3*h)
    E=vcat(E,transpose(E[i+1,:])+h*(1/6)*(k1+2*k2+2*k3+k4))
    t=vcat(t,[t[1,1]+(i+1)*h])
end
```

```julia
E
```

```julia
t
```

```julia
using Plots
```

```julia
plot(t,E[:,1],line=2,leg=false)
plot!(t,E[:,2],line=2,leg=false)
```
