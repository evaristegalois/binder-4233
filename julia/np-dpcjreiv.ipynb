<h4>Directional Derivative</h4>

Suppose that the temperature (in degrees Celsius) at the point $(x,y)$
near an airport is given by
\begin{equation}
  \label{eq:fthmegbk}
  f(x,y)=\frac{1}{180}\left(7400-4x-9y-0.03xy\right)\notag
\end{equation}
(with distances $x$ and $y$ measured in kilometres). Suppose that your
aircraft takes off from this airport at the location $P=(200,200)$ and
heads northeast in the direction specified by the vector $v=(3,4)$.
What initial rate of change of temperature will you observe?
