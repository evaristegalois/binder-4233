```julia
using DifferentialEquations
```

```julia
f(u,p,t)=t*u^3-t*u
```

```julia
u0=1
```

```julia
tspan=(0,1.2)
```

```julia
prob = ODEProblem(f,u0,tspan)
```

```julia
sol = solve(prob, Tsit5(), reltol=1e-8, abstol=1e-8)
```

```julia
using Plots
```

```julia
plot(sol,linewidth=3,xaxis="Time (t)",yaxis="u(t)",label="Julia's solution")
```

```julia
f(u,p,t)=(2*t*u+2*t)/(t^2+1)
```

```julia
u0=0
```

```julia
tspan=(0,7)
```

```julia
prob=ODEProblem(f,u0,tspan)
```
