```julia
using Roots
```

```julia
s(t)=-0.5*g*t^2+vo*t+so
```

```julia
s(5)
```

```julia
rr=0.04
```

```julia
C=49+(g/rr)
```

```julia
v(t)=(-g/rr)+C*exp(-rr*t)
```

```julia
v(5)
```

```julia
p=Gadfly.plot([t->v(t),t->0],-0.5,6)
```

```julia
tmax=(-1/rr)*log(g/(rr*C))
```

```julia
D=C/rr
```

```julia
s(t)=(-g/rr)*t+C*exp(-rr*t)*(-1/rr)+D
```

```julia
s(0)
```

```julia
s(tmax)
```

```julia
find_zero(s,10)
```
