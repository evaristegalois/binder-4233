<h2>Simpson's Rule</h2>

As you can appreciate now, there are integrals that are very difficult to solve. Perhaps they cannot be solved at all. In this case, we can use an approximation called Simpson's Rule. The formula is 

$$
\int_{a}^{b}f(x)\,dx\approx\frac{b-a}{6}\left(f(a)+4\cdot{}f\left(\frac{a+b}{2}\right)+f(b)\right)
$$

Simpson's Rule stems from replacing the function with the parabola that goes through $f(a),f\left(\frac{a+b}{2}\right),f(b)$. Let us try this for $f(x)=e^{-x^{2}}$ between $x=-\frac{1}{2}$ and $x=1$.

```julia
using Pkg
```

```julia
using Plots
```

```julia
x=range(-1.5,stop=1.5,length=100)
```

```julia
plot(x,exp.((-1).*x.^2),line=2,leg=false)
plot!([-1/2],[exp(-1/4)],seriestype= :scatter,mc= :blue)
plot!([1/4],[exp(-1/16)],seriestype= :scatter,mc= :blue)
plot!([1],[exp(-1)],seriestype= :scatter,mc= :blue)
plot!(range(-1.5,stop=1.5,length=100),-0.650796.*x.^2+0.051451.*x.+0.967225,line=2,leg=false)
```
